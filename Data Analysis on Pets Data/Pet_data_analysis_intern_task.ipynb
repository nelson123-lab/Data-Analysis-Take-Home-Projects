## How many owners have at least two different kinds of pets (e.g. a dog and a cat)?


```python
# keeping a dictionary for the different categories of breed by owner
cate_by_owner = {}
owner_count = 0

# Iterating throught the dataframe
for index, row in pets.iterrows():
    owner_id = row['OwnerID']
    category = row['Kind']

    # Checking whether the owner_id is present in the category dictionary or not and creating a set.
    if owner_id not in cate_by_owner:
      cate_by_owner[owner_id] = set()

    # Adding up the categories without duplicates
    cate_by_owner[owner_id].add(category)

# Checking for owners with different types of pets.
for owner_id, categories in cate_by_owner.items():
  if len(categories) >= 2:
    owner_count += 1
  else:
    pass
print(f"Owners having at least two kinds of pets {owner_count}")
```

## How many owners have pets where the first letter of their name (OwnerName) matches the first letter of their pet's name (PetName)? E.g. Cookie and Charles.

```python
owner_id = dict(zip(owners['OwnerID'],owners['Name'] + " " + owners['Surname']))
owner_id
```

```python
pets["Owner Name"] = pets["OwnerID"].apply(lambda x: pd.Series(x).map(owner_id))
pets
```

```python
Count = 0
for index, row in pets.iterrows():
  pet_name = row['Name']
  owner_name = row['Owner Name']

  if pet_name[0] == owner_name[0]:
    Count += 1
  else: pass
print(f"The number of owners having same starting letter as the pets name = {Count}")
```

## What percentage of cities have more male pets than female pets?

```python
owner_city = dict(zip(owners['OwnerID'], owners['City']))
pets["Owner City"] = pets["OwnerID"].apply(lambda x: pd.Series(x).map(owner_city))
pets.head(10)
```

```python
grouped = pets.groupby(['Owner City', 'Gender']).size().unstack()
grouped.fillna(0, inplace = True)
```

```python
grouped.reset_index()
```

### There are 40 cities in the data.

```python
male_dominans_count = 0
for index, rows in grouped.iterrows():
  female_count = rows['female']
  male_count = rows['male']

  if male_count > female_count:
    male_dominans_count += 1
  else: pass
print(f'The percentage of cities where male pets than female is {(male_dominans_count/grouped.shape[0])*100}')
```
